```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

from naive_bayes import naive_bayes
```

# Iris Dataset

## Overview

The sklearn iris dataset is a widely used benchmark dataset in machine learning. It is included in the scikit-learn library (sklearn) and consists of 150 samples of iris flowers, each belonging to one of three species: Setosa, Versicolor, and Virginica.


## Features

The dataset contains four features (in centimeter):

1. Sepal length 
2. Sepal width 
3. Petal length 
4. Petal width 

Each sample is labeled with the species it belongs to. This dataset is often used for classification tasks, where the goal is to predict the species of iris flower based on its features.


## Target

0. Setosa
1. Versicolor 
2. Virginica

```python
iris = load_iris()
X, y = iris.data, iris.target

print('input shape:', np.shape(X))
print('target shape:', np.shape(y))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

print('train input shape:', np.shape(X_train))
print('train target shape:', np.shape(y_train))
print()
print('test input shape:', np.shape(X_test))
print('test target shape:', np.shape(y_test))
```

### 2D Data visualization

```python
plt.figure(figsize=(12,6))
plt.tight_layout()
plt.xlabel("X0")
plt.ylabel("X2")
plt.scatter(X_train[:,0],X_train[:,2],cmap='bwr', c=y_train, edgecolor='k', s=40)
plt.title('2 features of the Dataset')
plt.show()
```

# Training

```python
model = naive_bayes()
```

```python
model.fit(X_train, y_train)
```

# Evaluation

```python
acc = model.evaluate(X_test, y_test)
print('Accuracy: ', acc)
```
