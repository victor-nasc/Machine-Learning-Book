```python
import matplotlib.pyplot as plt
import numpy as np

from kmeans import KMeans
%matplotlib inline
```

## Creating a random training set

```python
N = 75 # number of points

X = np.vstack([np.random.randn(N, 2) + [4, 4],
               np.random.randn(N, 2) + [-3, 0],
               np.random.randn(N, 2) + [3, -3]])
```

```python
plt.scatter(X[:, 0], X[:, 1], alpha=0.8)
plt.title('Dataset')
plt.show()
```

## Clustering

```python
kmeans = KMeans(k=3)
kmeans.fit(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels, cmap='viridis', alpha=0.8)
plt.scatter(kmeans.centroids[:, 0], kmeans.centroids[:, 1], c='red', s=100, alpha=0.8)
plt.title('KMeans Clustering')
plt.show()
```
